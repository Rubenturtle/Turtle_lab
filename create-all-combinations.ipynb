```python
import pandas as pd
import itertools
import os
```

```python
def get_all_csv_file_list(path):
    File_list = [] #f for f in os.listdir(path) if os.isfile(mypath,f)
    for file in os.listdir(path):
        if file.endswith(".csv"):
            if file not in File_list:
                File_list.append(os.path.join(path,file))
        else:
            pass
    return File_list

def create_folder_if_not_exists(folder_path):
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)
        print(f"Folder created at: {folder_path}")
    else:
        print(f"Folder already exists at: {folder_path}")
```

```python
input_path = r"Z:\data\im-nca-colombia\landcover\illogical_transitions"
output_path = input_path + r"\output_files"
file_list = get_all_csv_file_list(input_path)
create_folder_if_not_exists(output_path)
```

```python
# Create an empty DataFrame to store all results
full_df = pd.DataFrame(columns=['col1', 'col2'])

for file in file_list:
    df = pd.read_csv(file, header=None)

    # Get the column names (assuming there are exactly two columns)
    col1 = df[0]
    col2 = df[1]

    col1_str = col1.astype(str).str.strip() # Convert to string to handle all values
    col1_values = pd.to_numeric(col1_str, errors='coerce').tolist() # Convert to numeric values, coercing errors to NaN
    col1_values = [int(x) if pd.notna(x) else None for x in col1_values] # Convert NaNs to None and integers
    col1_values = [x for x in col1_values if x is not None] # Remove None values from the list

    col2_str = col2.astype(str).str.strip() # Convert to string to handle all values
    col2_values = pd.to_numeric(col2_str, errors='coerce').tolist() # Convert to numeric values, coercing errors to NaN
    col2_values = [int(x) if pd.notna(x) else None for x in col2_values] # Convert NaNs to None and integers
    col2_values = [x for x in col2_values if x is not None] # Remove None values from the list

    
    # Generate all possible combinations between non-empty values in the two columns
    combinations_list = list(itertools.product(col1_values, col2_values))

    filtered_combinations = [
        (v1, v2) for v1, v2 in combinations_list if not (pd.isna(v1) or pd.isna(v2))
    ]

    # Convert the list of combinations to a DataFrame
    result_df = pd.DataFrame(filtered_combinations, columns=['col1', 'col2'])

    # Append the new result_df into the full_df (stacking the rows)
    full_df = pd.concat([full_df, result_df], ignore_index=True)

# Write the result to a new CSV file
output_file = os.path.join(output_path, "illogical_transitions.csv")
full_df.to_csv(output_file, index=False, header=None)

print(f"Combinations have been saved to {output_file}")
```
