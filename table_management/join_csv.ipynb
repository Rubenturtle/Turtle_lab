```python
import pandas as pd
import glob
import os
```

```python
def merge_vcs_all_years(vcs_files):
    for file in vcs_files:
        vcs = pd.read_csv(file, sep=';')
        vcs = vcs.rename(columns={'Unnamed: 0':"cid"})
        vcs_copy = vcs.copy()
        vcs.drop(columns=['name'], inplace=True)
        try:
            vcs_df = pd.merge(vcs_df, vcs, on = ["cid"])
        except:
            vcs_df = pd.DataFrame(vcs)

    vcs_df[sorted(vcs_df)]

    vcs_name = vcs_copy["name"]
    vcs_df = pd.concat([vcs_name, vcs_df], axis=1)
    return vcs_df
```

```python
input_folder = r"C:\Users\ruben.crespo\Downloads\vegetation_carbon_stock_2001-2020"  # Update this with the path to your CSV files
output_file = os.path.join(input_folder,'combined_data.csv')    # The name of the output combined CSV file
```

```python
csv_files = glob.glob(os.path.join(input_folder, '*.csv'))

combined_df = merge_vcs_all_years(csv_files)

combined_df.to_csv(output_file, index=False, sep=',')
```
