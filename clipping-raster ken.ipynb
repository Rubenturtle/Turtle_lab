We are going to clip some rasters into a big mosaic. First is to locate all the files, and them one by one
do the clipping into a created folder of the same name.
This code was made so for ken in relation to the mosaic data

```python
#Import the libraries
import os
import sys
from osgeo import gdal
import numpy as np
import glob


gdal.UseExceptions()
```

```python
#Get a list of the files inside the folder
path = r"//akif.internal/public/z_resources/hydrosoil/hihydrosoil_ruben"
File_list = [] #f for f in os.listdir(path) if os.isfile(mypath,f)
for file in os.listdir(path):
    if ".tif" in file:
        if file not in File_list:
            File_list.append(file)
    else:
        pass
print(File_list)
```

```python
#we extract for now the properties of a file
"""We specify the current working direcroty"""
# os.chdir(r"//akif.internal/public/z_resources/hydrosoil/hihydrosoil_ruben") #works fine
os.chdir(r"//akif.internal/public/demo-geo-stack/geoserver/data/ken")
path = r"//akif.internal/public/demo-geo-stack/geoserver/data/ken"
"""We open the file"""
file_snap_list = []
# for file in File_list:
file = "nlcd_2019_land_cover_conus_reclassify.tif" #esto se omite cuando hagamos todos
datafile = gdal.Open( file )
dataname = file.replace('.tif','')

"""Here we create and go the folder which will contain the mosaic"""
dir = os.path.join(path, dataname + "_mosaic")
if not os.path.exists(dir):
    os.mkdir(dataname + "_mosaic")
# mosaic_path = os.open(dataname + "_mosaic")
os.chdir(dir)

"""Here we get the size of the file"""
width = datafile.RasterXSize #columns
height = datafile.RasterYSize #rows
# define your tile size
# it could be 256,512 or 1024. it depends on your scope
tilesize = 1024
"""Here we start with the tyiling"""
for i in range(0, width, tilesize): #tilesize marks from where to where in width
    for j in range(0, height, tilesize): #tilesize in height
        #for the edge parts, so we don't get nodata from the borders
        w = min(i+tilesize, width) - i
        h = min(j+tilesize, height) - j
        #we preapre the sentence
        output_file = dataname + "_" + str(i) + "_" + str(j) + ".tif"
        file_snap_list.append(output_file)
        try:
            ds = gdal.Translate(output_file, datafile, format = 'GTiff', 
            # noData = -32768,
            # outputType = gdal.GDT_Int16, 
            creationOptions = ['COMPRESS=DEFLATE', 'PREDICTOR=1']
            , srcWin = [str(i), str(j), str(w), str(h)])
            print(ds)
        except RuntimeError:
            print("The script got an error")
            sys.exit(1)
"""We have to close the file in order to proceed for the next parts"""
ds = None
    
```
