```python
import os
import rasterio
from rasterio import features
import geopandas as gpd
from shapely.geometry import mapping, box
from fiona.crs import from_epsg
from rasterio.crs import CRS
```

```python
path = r"C:\Users\admin\Downloads\global_forest_watch_tree_cover_gain_30m_2000_2012_mosaic"
output_shapefile = r"C:\Users\admin\Downloads\grid.shp"
```

```python
def get_raster_file_list(path):
    """Get a list of the raster files inside the folder"""
    File_list = [] #f for f in os.listdir(path) if os.isfile(mypath,f)
    for file in os.listdir(path):
        if file.endswith(".tif") or file.endswith(".tiff"):
            if file not in File_list:
                File_list.append(os.path.join(path,file))
        else:
            pass
    return File_list
```

```python
polygons = []
raster_list = get_raster_file_list(path)

# Get the CRS from the first raster (They are all the same)
with rasterio.open(raster_list[0]) as src:
    crs = CRS.from_dict(src.crs)

counter = 0
for raster in raster_list:
    print()
    with rasterio.open(raster) as src:
        bbox = box(*src.bounds)
        polygons.append(bbox)

gdf = gpd.GeoDataFrame(geometry=polygons)

# Set the CRS of the GeoDataFrame
#Manually
# crs = from_epsg(4326)  # WGS84
# gdf.crs = crs

#From the data source
gdf.crs = crs
```
